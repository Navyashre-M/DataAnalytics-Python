# Encoding:
 - converting categorical data into numerical data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.read_csv(r"C:\My python Files\Salary_EDA.csv")
df.head()
```

Filter Categorial features

```python
categorical_column=['Education Level']
```

Define and Apply Encoder

```python
encoder=OneHotEncoder(drop=None,sparse_output=False)
```

```python
encoded_data=encoder.fit_transform(df[categorical_column])
```

```python
print(encoded_data)
```

The encoded data is in the form of Array.Now,we need to convert the encoded dataframe with categorical as column names

```python
encoded_df=pd.DataFrame(encoded_data , columns=encoder.get_feature_names_out(categorical_column))
encoded_df.head()
```

```python
encoded_df.drop('Education Level_nan',axis=1)
```

```python
encoded_df.drop('Education Level_nan',axis=1,inplace=True)
encoded_df.head()
```

```python
final_df=pd.concat([df,encoded_df],axis=1)
final_df.head()
```
